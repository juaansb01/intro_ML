#Fases de un proyecto de Data Science con Machine Learning

Teniendo en cuenta lo comentado en la guía de introducción vista anteriormente, veamos ahora las distintas fases que encontramos al trabajar en un proyecto de Data Science sobre el cual queremos aplicar distintos modelos de Machine Learning (ML).

1. **Recopilación de datos.**
>En esta fase se recolectan y obtienen los datos necesarios para el proyecto de ML. Esto puede implicar la recopilación de datos de diversas fuentes, como bases de datos existentes, sensores, registros históricos o datos generados por usuarios.
2. **Preprocesamiento de datos.**
>Los datos suelen estar en bruto y pueden contener ruido, valores faltantes o inconsistentes. En esta etapa, se llevan a cabo tareas como limpieza de datos, eliminación de valores atípicos, normalización y transformación para preparar los datos para el modelado.
3. **Exploración de datos.**
>Es necesario llevar a cabo análisis exploratorios para comprender mejor los datos. Esto incluye la visualización de datos, el cálculo de estadísticas descriptivas y la identificación de relaciones o patrones significativos.
4. **Selección de características.**
>Las características utilizadas para entrenar el modelo son críticas. En esta fase se seleccionan las características más relevantes, y en ocasiones, se crean nuevas características a partir de las existentes.
5. **División de datos.**
>Los datos se dividen en conjuntos de entrenamiento y prueba para evaluar el rendimiento del modelo. Esto es esencial para evitar el sobreajuste del modelo a los datos de entrenamiento.
6. **Selección de distintos algoritmos de Machine Learning.**
>Dependiendo de las características de los datos, elegimos varios modelos de Machine Learning que sean adecuados para el problema, teniendo siempre en mente el objetivo del proyecto. Los entrenamos y comparamos su rendimiento sobre un conjunto de validación, a través de distintas métricas.
7. **Validación del modelo.**
>Se evalúan los modelos en el conjunto de prueba para medir su rendimiento en datos no vistos. De nuevo, utilizaremos métricas de evaluación para determinar cómo de bien funcionan.
8. **Ajuste de hiperparámetros.**
>En esta fase, se llevan a cabo los ajustes necesarios de los hiperparámetros de los distintos modelos para mejorar su rendimiento. Esto puede incluir la optimización de la tasa de aprendizaje, la profundidad del modelo, elección del kernel, etc., dependiendo del modelo. Esta fase es intercambiable con la fase anterior y recursiva, pues se puede realizar tantas veces como necesitemos para mejorar el rendimiento de los modelos lo máximo posible. También, si fuera necesario, podríamos retocar desde la fase 2 del proyecto, para incrementar de nuevo la eficiencia de los modelos.
9. **Elección del modelo.**
>Gracias a las distintas mediciones de rendimiento de los modelos que hemos llevado a cabo en las etapas previas, basándonos en el objetivo final del proyecto, decidimos el mejor modelo posible para resolver el problema al que nos enfrentamos.
10. **Consideraciones a partir de nuestro análisis.**
>En caso de necesitar más información, continuaremos el análisis y la evaluación del modelo para obtener mejores resultados.
11. **Despliegue del modelo.**
>Una vez que el modelo satisface los requisitos de rendimiento, se puede desplegar para que lleve a cabo predicciones o tome decisiones en tiempo real.
12. **Monitorización y mantenimiento.**
>Por último, se debe monitorizar el funcionamiento del modelo para asegurarse que lo hace correctamente. Si fuera necesario, podríamos realizar actualizaciones y mantenimiento continuo.
